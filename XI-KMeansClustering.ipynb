# K-Means Clustering

 Very useful in high dimensional data
- **INTUITION**
- Decide k number of clusters
- 1st randomly select k centeroids
- Assign each datapoint to centroids based on least ecludian distance between centroid & datapoint & form a cluster
- From the newly formed k clusters,will calculate centroid by calcualting mean values, & redo above mentioned point
- Move centroid (doing the above mentioned point till we no long find new centroid)
- We no long find new centroids
- We will use Elbow method to determine value of K

**WCSS(Within Cluster Sum of Squares) - WCSS is the sum of the squared distance between each point and the centroid in a cluster**

- To determine number of clusters that is K 
- We will start increasing the value of k from 1 & keep calculating WCSS & plotting them 
- For each increase in the number of cluster we will add WCSS for each clustet to plot
- We will keep adding clusters till the slope of the curve stablises or slope not changing much
- For each addition of cluster the values of WCSS will keep decreasing
- WCSS1>WCSS2>WCSS3...


```python
from sklearn.datasets import make_blobs
# Generate isotropic Gaussian blobs for clustering.
import matplotlib.pyplot as plt
```
